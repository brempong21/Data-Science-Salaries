```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/brempong21/Data-Science-Salaries/main/data_science_salaries.csv')
df.head(10)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
plt.hist(df['work_year'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Work Years')
plt.xlabel('Work Year')
plt.ylabel('salary_in_usd')
plt.show()
```

```python
sns.countplot(x='experience_level', data=df)
plt.title('Count of Experience Levels')
plt.xlabel('Experience Level')
plt.ylabel('Count')
plt.show()
```

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
agg_data = df.groupby('company_size')['salary_in_usd'].mean().reset_index()
sns.barplot(x='company_size', y='salary_in_usd', data=agg_data)
plt.title('Average Salary by Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary in USD')
plt.show()
```

```python
sns.kdeplot(df['work_year'], fill=True, color='skyblue')
plt.title('Kernel Density Estimation (KDE) of Salary')
plt.xlabel('Salary in USD')
plt.ylabel('Density')
plt.show()
```

```python
experience_counts = df['experience_level'].value_counts()
plt.pie(experience_counts, labels=experience_counts.index, autopct='%1.1f%%', startangle=90, colors=['lightcoral', 'lightskyblue', 'lightgreen'])
plt.title('Distribution of Experience Levels')
plt.show()
```
